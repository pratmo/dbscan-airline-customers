```python
os.chdir("/home/prathikm/Desktop/ds_assignments/doing/clustering")
df_n = pd.read_excel("EastWestAirlines.xlsx") #import airline dataset
df_n
```

# Drop the column ID - it is of no use

```python
df = df_n.drop(['ID#'],axis=1)
df
```

```python
array=df.values
```

```python
array #DBSCAN expects data in the form of an array/numpy representation
```

```python
stscaler = StandardScaler().fit(array) #stadardize the data
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=1.8, min_samples=12) #Min = D+1 where D is the number of columns.
#as we decrease eps, more outliers are formed as radius decreases.
dbscan.fit(X)
```

```python
#Labels are nothing but the clusters. #Noisy samples are given the label -1. 
dbscan.labels_ #0 means normal data-point of cluster, -1 means noisy datapoint

#Here algorithm is only forming one cluster as 0 is only formed. So DBSCAN may not be the right approach for 
#this dataset.
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pf = pd.concat([df,cl],axis=1)
```

```python
pf
```
